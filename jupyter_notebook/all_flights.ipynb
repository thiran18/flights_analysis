```python
import pandas as pd
import requests
```

```python
API_KEY="b6584cbf-6739-4d78-9208-fc98f30f13d0"
AIRLINES = ["6E","AI","IX","QP","SG","9I","S5","IC"]
```

```python
all_flights = []

for code in AIRLINES:
    url = f"https://airlabs.co/api/v9/schedules?airline_iata={code}&api_key={API_KEY}"
    response = requests.get(url)
    data = response.json()
    flights = data.get("response", [])
    print(f"Fetched {len(flights)} flights for {code}")
    all_flights.extend(flights)
```

```python
print("Total flights fetched:",len(all_flights))
```

```python
flights=pd.DataFrame(all_flights)
```

```python
flights['dep_icao'] = flights['dep_icao'].fillna('Unknown')
flights['arr_icao'] = flights['arr_icao'].fillna('Unknown')
flights['duration'] = flights['duration'].fillna(flights['duration'].median())
```

```python
flights['dep_estimated'] = flights['dep_estimated'].fillna(flights['dep_time'])
flights['dep_estimated_utc'] = flights['dep_estimated_utc'].fillna(flights['dep_time_utc'])

flights['arr_estimated'] = flights['arr_estimated'].fillna(flights['arr_time'])
flights['arr_estimated_utc'] = flights['arr_estimated_utc'].fillna(flights['arr_time_utc'])

flights['dep_actual'] = flights['dep_actual'].fillna(flights['dep_time'])
flights['dep_actual_utc'] = flights['dep_actual_utc'].fillna(flights['dep_time_utc'])

flights['arr_actual'] = flights['arr_actual'].fillna(flights['arr_time'])
flights['arr_actual_utc'] = flights['arr_actual_utc'].fillna(flights['arr_time_utc'])
```

```python
text_columns = ['dep_terminal','dep_gate','arr_terminal','arr_gate','arr_baggage',
                'cs_airline_iata','cs_flight_number','cs_flight_iata','aircraft_icao']

numeric_columns = ['delayed','dep_delayed','arr_delayed']

flights[text_columns] = flights[text_columns].fillna('Unknown')
flights[numeric_columns] = flights[numeric_columns].fillna(0)
```

```python
time_columns = ['dep_time','arr_time','dep_actual','arr_actual','dep_estimated','arr_estimated']
for col in time_columns:
    flights[col] = pd.to_datetime(flights[col])
```
